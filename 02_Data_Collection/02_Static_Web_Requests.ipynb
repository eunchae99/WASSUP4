## 멜론차트, 노래제목과 가수 수집하기
```
https://www.melon.com/chart/index.htm

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

````python
PATH = 'https://www.melon.com/chart/index.htm'
resp = requests.get(PATH)
resp
````

````python
resp.text
````

````python
info = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}
info
````

````python
resp = requests.get(PATH, headers = info)
resp
````

````python
html_src = resp.text
html_src
````

````python
soup = BeautifulSoup(html_src, 'lxml')
soup
````

## 네이버 검색결과 수집하기
```
쿼리를 사용자입력받아서 수집하기

https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=ai

````python
PATH = 'https://search.naver.com/search.naver'
PATH
````

````python
keyword = input('검색어를 입력하세요 : ')
````

````python
resp = requests.get(PATH, params={'query':keyword})
resp
````

````python
html_src = resp.text
html_src
````

## Quiz: 타이틀과 url 출력

````python
soup = BeautifulSoup(html_src, 'lxml')
soup
````

````python
result = soup.select('.news_tit')
print(len(result))
print(result)
````

````python
for a in result:
    print(a.text, a['href'])
````
